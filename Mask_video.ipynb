```python
import cv2
import numpy as np
import glob
import os
```

```python
def mask_video(video_path, mask_path, output_path):

    # Open video
    cap = cv2.VideoCapture(video_path)

    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Output writer
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # Load mask
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    mask = cv2.resize(mask, (width, height))

    # Ensure binary mask
    _, mask = cv2.threshold(mask, 127, 255, cv2.THRESH_BINARY)
    
    # Invert mask
    mask = cv2.bitwise_not(mask)

    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Apply mask
        result = cv2.bitwise_and(frame, frame, mask=mask)

        out.write(result)
        frame_count += 1

    cap.release()
    out.release()

    print(f"Finished processing {frame_count} frames.")
    print(f"Output saved to: {output_path}")
```

```python
# Process one video

vid_nr = "018"

video_path = f"/home/jovyan/M2-2_stage/ICG_video's_filtered/FIM{vid_nr}_ICG_filtered.mp4"
mask_path = f"/home/jovyan/M2-2_stage/Masks/FIM{vid_nr}_mask.jpg"
output_path = f"/home/jovyan/M2-2_stage/Masked_videos/FIM{vid_nr}_ICG_filtered_masked.mp4"

mask_video(video_path, mask_path, output_path)
```

```python
# Loop over all videos
video_files = glob.glob("/home/jovyan/M2-2_stage/ICG_video's_filtered/FIM*_ICG_filtered.mp4")

for video_path in video_files:
    
    filename = os.path.basename(video_path)  
    vid_id = filename.split("_")[0]   # e.g. FIM014
    
    mask_path = f"/home/jovyan/M2-2_stage/Masks/{vid_id}_mask.jpg"
    output_path = f"/home/jovyan/M2-2_stage/Masked_videos/{vid_id}_ICG_filtered_masked.mp4"
    
    print(f"Processing {vid_id}")
    
    if os.path.exists(mask_path):
        mask_video(video_path, mask_path, output_path)
    else:
        print(f"Mask missing for {vid_id}")
```
